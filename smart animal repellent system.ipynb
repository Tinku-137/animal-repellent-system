```python
data_yaml = """
train: /content/drive/MyDrive/Smart Animal Repellent System/train
val: /content/drive/MyDrive/Smart Animal Repellent System/test

nc: 14
names: ['sheep', 'snake', 'pig', 'rabbit', 'monkey', 'goat', 'elephant', 'deer', 'cattle', 'bull', 'chicken', 'bear', 'sparrow', 'parrot']
"""

with open("/content/drive/MyDrive/Smart Animal Repellent System/data.yaml", "w") as f:
    f.write(data_yaml)

print("‚úÖ data.yaml file created/updated successfully!")
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # or your preferred YOLO model

model.train(
    data='/content/drive/MyDrive/Smart Animal Repellent System/data.yaml',
    epochs=50,
    imgsz=640,
    device='cpu',   # üëà Force CPU
    project='/content/drive/MyDrive/Smart Animal Repellent System/runs',
    name='train_cpu'
)
```
