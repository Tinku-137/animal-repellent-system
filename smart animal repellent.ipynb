```python
import yaml

# Path to your data.yaml
yaml_path = '/content/drive/MyDrive/Smart Animal Repellent System/data.yaml'

# Load YAML file
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Print basic info
print("âœ… Dataset Info:")
print(f"Train Path: {data.get('train')}")
print(f"Validation Path: {data.get('val')}")
print(f"Total Classes (nc): {data.get('nc')}\n")

# Print all class names
print("ğŸ“¦ Classes Present in Dataset:")
for i, name in enumerate(data.get('names', [])):
    print(f"{i}: {name}")
```

```python
import os
import shutil

# Define dataset paths
base_dir = "/content/drive/MyDrive/Smart Animal Repellent System"
train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")

# Animals to keep (based on your levels)
animals_to_keep = [
    "cow", "buffalo", "elephant", "monkey",
    "pig", "dog", "deer", "boar",
    "bird", "rabbit"
]

# Function to clean dataset
def clean_dataset(folder):
    for animal in os.listdir(folder):
        path = os.path.join(folder, animal)
        if os.path.isdir(path) and animal not in animals_to_keep:
            print(f"ğŸ—‘ï¸ Removing unwanted animal folder: {animal}")
            shutil.rmtree(path)

# Clean train and test folders
clean_dataset(train_dir)
clean_dataset(test_dir)

print("âœ… Dataset cleaned successfully! Only selected animal folders remain.")
```

```python
data_yaml_content = """
train: /content/drive/MyDrive/Smart Animal Repellent System/train
val: /content/drive/MyDrive/Smart Animal Repellent System/test

nc: 10
names: ['cow', 'buffalo', 'elephant', 'monkey', 'pig', 'dog', 'deer', 'boar', 'bird', 'rabbit']
"""

# Save it to a file
with open("/content/drive/MyDrive/Smart Animal Repellent System/data.yaml", "w") as f:
    f.write(data_yaml_content)

print("âœ… data.yaml file created successfully!")
```
